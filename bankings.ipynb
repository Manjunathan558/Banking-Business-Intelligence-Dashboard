```python
query = "select * from banking_case.customer"
```

```python
df = pd.read_sql(query, cnx)
```

```python
bins =[0,100000,300000,float('inf')]
labels = ['Low','Med','High']
df['Income Band']=pd.cut(df['Estimated Income'],bins=bins, labels=labels,right= False)

```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
cols = df[["BRId","GenderId","IAId","Amount of Credit Cards", "Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns

for col in cols:
    print(f"Value counts for '{col}':")
    display(df[col].value_counts())
```

```python
for i, predictor in enumerate(df[["BRId","IAId","Amount of Credit Cards", "Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting","Income Band"]].columns):
    plt.figure(i)
    sns.countplot(data=df,x=predictor,hue='Nationality')
    plt.xticks(rotation=45)
    plt.tight_layout()
```

```python
for col in cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()
```

```python
numerical_cols = ['Estimated Income' , 'Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits', 'Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

plt.figure(figsize=(10,15))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

```python
correlation_matrix= df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title("correlation_matrix")
plt.show()
```
